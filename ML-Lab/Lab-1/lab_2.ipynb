```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
fig.add_axes([0, 0, 1, 1])
x = np.linspace(0, 50, 10)
y = np.linspace(0, 200, 10)
plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Plotting")
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([.2, .5, .2, .2])
ax1.plot(x, y)
ax2.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Plotting")
plt.show()
```

```python
df = pd.read_csv("company_sales_data.csv")
df.head(5)
```

```python
data_df = df.loc[:, ['month_number', 'total_profit']]
data_df
```

```python
fig = plt.figure()
fig.add_axes([0, 0, 1, 1])
plt.plot('month_number', 'total_profit', data=data_df)
plt.xlabel("Month number")
plt.ylabel("Total profit")
plt.title("Profit Monthwise")
plt.show()
```

```python
fig = plt.figure()
fig.add_axes([0, 0, 1, 1])
line = plt.plot(df['month_number'], df['total_units'], 'r:', linewidth=3)
line = line[0]
plt.xlabel("Month number")
plt.ylabel("Sold units number")
plt.title("Units Sold Monthwise")
plt.legend("Graph", loc='lower right')
line.set_marker('o')
line.set_markerfacecolor('red')
line.set_markeredgecolor('red')
line.set_linewidth(3)
plt.show()
```

```python
cols = list(df.columns)[1:7]
x = df['month_number']
fig = plt.figure()
fig.add_axes([0, 0, 1, 1])
for col in cols:
    plt.plot(x, df[col], label=col)
plt.xlabel("Month Number")
plt.ylabel("Units sold")
plt.title("Units sold per month for each product")
plt.legend(loc="upper left")
plt.show()
```

```python
def total_in_year(ser):
    return ser.sum()
```

```python
ser = df.loc[:, cols].apply(lambda x: total_in_year(x), axis=0)
plt.pie(ser, labels=cols)
plt.title("Split up of each product in sales")
plt.show()
```
